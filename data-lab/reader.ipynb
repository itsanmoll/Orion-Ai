```python
import pandas as pd 
import numpy as np
import datetime
```

```python
def read_solar(PATH: str) -> pd.DataFrame:
    cols=['YYY','MM','DD','days','days_m','Bsr','dB',
                'Kp1','Kp2','Kp3','Kp4','Kp5','Kp6','Kp7','Kp8',
                'ap1','ap2','ap3','ap4','ap5','ap6','ap7','ap8','Ap',
                'SN','F10.7obs','F10.7adj','D']
    df = pd.read_csv(PATH, sep=r'\s+', header=None, names=cols)
    df.columns = [a.lower() for a in df.columns]
    df['epoch'] = df.apply(lambda row: datetime.datetime(int(row.yyy), int(row.mm), int(row.dd)), axis=1)
    df.drop(['yyy','mm','dd','days','days_m'], axis=1, inplace=True)
    df['f10.7_81'] = None
    for i in range(0,len(df)-80,81):
        df.loc[np.arange(i,i+81),'f10.7_81'] = df.loc[np.arange(i,i+81),'f10.7obs'].mean()
    return df
```

```python
PATH1 = 'kpdata3yr.txt'
```

```python
solar = read_solar(PATH1)
solar
```

```python
solar.to_csv("kpcsv-3yr.csv", index=False)
```
